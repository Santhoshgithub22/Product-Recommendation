```python
full_sales_df = pd.concat([sales_2017_df,sales_2018_df,sales_2019_df], ignore_index=True, sort=False)
```

```python
full_sales_df
```

(i) Drop the Gross, Disc, Voucher Amount columns

```python
sales_df=full_sales_df.drop(columns=["Gross","Disc","Voucher Amount"], axis=1)
```

```python
sales_df
```

(ii) Removing null rows

```python
sales_df
```

(iii) Convert the date from DD/MM/YY to YY/MM/DD 

```python
sales_df['Date'] = pd.to_datetime(sales_df['Date'], dayfirst=True)
```

```python
sales_df
```

(iv) Removing the Sal string in voucher column using .str.slice(star,stop,step)method

```python
sales_df["Voucher"]=sales_df["Voucher"].str.slice(start=4, stop=None, step=1).astype(int)
```

```python
sales_df.head(5)
```

(v) Convert the party column and product column into upper case

```python
sales_df["Party"]=sales_df["Party"].str.upper()
```

```python
sales_df["Product"]=sales_df["Product"].str.upper()
```

```python
sales_df.head(5)
```

(vi) Remove comma in quantity column

```python
sales_df["Qty"]=sales_df["Qty"].str.replace(',','').astype("float").astype(int)
```

```python
sales_df.head(5)
```

(vi) Remove comma in rate column

```python
sales_df["Rate"]=sales_df["Rate"].str.replace(',','').astype(float)
```

```python
sales_df.head()
```

```python
sales_df = sales_df.sort_values(['Date','Voucher'])
```

```python
sales_df
```

(vii) Save into CSV file

# 2. Product Ranking

# Finding top selling items

```python
top_sell_items_df = sales_df.groupby("Product").agg({'Qty':'sum', 'Rate':'max'})
```

```python
top_sell_items_df
```

(iii) Reset the index by converting product column into data frame column

```python
top_sell_items_df.reset_index(inplace=True)
```

```python
top_sell_items_df
```

(iv) Rank the product by most quantity sold

```python
top_sell_items_df['Top_Sell_Rank']=top_sell_items_df['Qty'].rank(method='min')
```

```python
top_sell_items_df
```

```python
top_sell_items_df['Top_Sell_Rank']=top_sell_items_df['Qty'].rank(method='min', ascending=False).astype(int)
```

```python
top_sell_items_df
```

```python
top_sell_items_df.sort_values('Qty', ascending=False)
```

(v) Top 20 items sold

```python
top_sell_items_df.sort_values('Qty', ascending=False).head(20)
```

# Most popular items

(i) Remove duplicate record in product, date, and party columns

```python
unique_order_items_df = sales_df.drop_duplicates(["Product", "Date", "Party"])
```

```python
unique_order_items_df
```

(ii) Find the number of orders placed and the unique no. of customers placed orders of each product

```python
most_popular_items_df = unique_order_items_df.groupby('Product').agg({'Date':'count', 'Party':'nunique'})
```

```python
most_popular_items_df
```

```python
most_popular_items_df.columns=['No_of_Orders', 'No_of_Customers']
```

```python
most_popular_items_df
```

(iii) Reset the index by converting the product into the column

```python
most_popular_items_df.reset_index(inplace=True)
```

```python
most_popular_items_df
```

(iv) Most Frequently Purchased Items

Product with high number of orders can be considered as the most frequently purchased items.

To find the most popular items, include the number of customers purchased and provide more weightage to products purchased by more customers

Weighted No. of Orders (W) = O * ( C / M )

where,

O = No. of Orders

C = No. of Customers purchased the product

M = Maximum number of customers made transaction in the entire period

```python
O = most_popular_items_df["No_of_Orders"]
C = most_popular_items_df["No_of_Customers"]
M = most_popular_items_df["No_of_Customers"].max()
```

```python
M
```

```python
most_popular_items_df["Weighted_No_of_Orders"] = O * ( C / M )
```

```python
most_popular_items_df
```

(v) Rank the products by weighted number of orders

```python
most_popular_items_df["Popularity_Rank"]=most_popular_items_df["Weighted_No_of_Orders"].rank(method='min', ascending=False).astype(int)
```

```python
most_popular_items_df
```

(vi) Top 20 most popular items sold

```python
most_popular_items_df.sort_values('Popularity_Rank', ascending=True).head(20)
```

(vii) Merge all the ranks

```python
product_rankings_df=pd.merge(top_sell_items_df, most_popular_items_df, how='inner', on='Product')
```

```python
product_rankings_df
```

(vii) Getting only Product, Price and Rank Columns

```python
product_rankings_df=product_rankings_df[["Product", "Rate", "Top_Sell_Rank", "Popularity_Rank"]]
```

```python
product_rankings_df
```

(viii) Top 20 product rankings

```python
product_rankings_df.sort_values('Popularity_Rank', ascending=True).head(20)
```

Save a product ranking into CSV

```python
product_rankings_df.to_csv('Product-Rankings.csv', index=False)
```

Create a pickle file with ranking data frame

```python
pickle.dump(product_rankings_df, open('Prod_ranking_model.pkl','wb'))
```

# 4. Recommend Product To Customer

```python
sales_df = pd.read_csv("Sales-Transactions-Edited.csv")
```

Build correlation matrix for the customer product relation (using user-user based relation recommendation)

```python
prod_cust_qty_df = sales_df.groupby(['Product','Party']).agg({'Qty' : 'sum'})
```

```python
prod_cust_qty_df
```

```python
prod_cust_qty_df.reset_index(inplace=True)
```

(i) Find the number of unique customers purchased each product

```python
prod_cust_count_df = sales_df.groupby(['Product']).agg({'Party':'nunique'})
prod_cust_count_df.columns=['No_of_Customers']
```

```python
prod_cust_count_df.reset_index(inplace=True)
prod_cust_count_df
```

(ii) Merge the unique customer count and quantity purchased of each product

```python
prod_cust_df=pd.merge(prod_cust_qty_df, prod_cust_count_df, how='inner', on='Product')
```

```python
prod_cust_df
```

(iii) Create a pivot table with all customers on columns, products on rows, and quantity as values

```python
prod_cust_pivot_df = prod_cust_df.pivot(index='Product', columns='Party', values="Qty").fillna(0)
```

```python
prod_cust_pivot_df
```

(iv) Find the correlation between every two customers and build a correlation matrix using corr() method

Using Spearman method to identify the correlation. Pearson was not providing the better results and kendall is taking the long time for execution

```python
cust_correlation_df = prod_cust_pivot_df.corr(method='spearman', min_periods=5)
```

```python
cust_correlation_df
```

Saving a customer to customer correlation matrix to a CSV model

```python
cust_correlation_df.to_csv('Customer-Customer-Correlation-Matrix.csv')
```

Saving as a pickle file

```python
pickle.dump(cust_correlation_df, open('cust_correlation_model.pkl', 'wb'))
```
